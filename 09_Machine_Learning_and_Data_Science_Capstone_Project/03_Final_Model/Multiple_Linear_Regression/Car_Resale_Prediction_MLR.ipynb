# Importing Libraries and Data Preprocessing

```python
import pandas as pd
```

```python
df=pd.read_csv('car data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
finaldf=df[['Year','Selling_Price','Present_Price','Kms_Driven','Fuel_Type','Seller_Type','Transmission','Owner']]
```

```python
finaldf.head()
```

```python
finaldf['currentyear']=2021
finaldf.head()
```

```python
finaldf['years']=finaldf['currentyear']-finaldf['Year']
finaldf.head()
```

```python
finaldf.head()
```

```python
finaldf.drop(['Year','currentyear'],axis=1,inplace=True)
```

```python
finaldf=pd.get_dummies(finaldf,drop_first=True)
```

```python
finaldf.head()
```

```python
finaldf.to_csv("Pre_car_data.csv",index=True)
```

```python
x=finaldf.drop('Selling_Price',axis=1)
```

```python
y=finaldf['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x,y)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
